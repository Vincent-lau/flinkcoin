```python
import matplotlib.pyplot as plt
import numpy as np
import json
from palettable.colorbrewer.qualitative import Set2_7
import matplotlib.font_manager as font_manager
```

```python
params = {
   'axes.labelsize': 16,
   'font.size': 14,
   'legend.fontsize': 18,
   'xtick.labelsize': 14,
   'ytick.labelsize': 14,
   'text.usetex': False,
   'figure.figsize': [8, 5],
   'hist.bins': 20,
}
rcParams = plt.rcParams
rcParams.update(params)
graph_dir = "graphs/"
cs = Set2_7.mpl_colors
```

```python
x = []
y_price = []
y_pred = []
y_rmse = []
```

```python

with open('log/output3.txt', 'r') as f:
  for line in f:
    data = json.loads(line)
    x.append(data['counter'])
    y_price.append(data['actual_buy_mean'])
    y_pred.append(data['buy_mean'])
    y_rmse.append(data['rmse'])

```

```python
fig, ax = plt.subplots()
# fig.set_size_inches((16, 9))
ax.plot(x, y_price, label='actual price', color=cs[0])
ax.plot(x, y_pred, label='prediction', color=cs[1])

ax.set_xlabel('Time points')
ax.set_ylabel('Price')
ax.legend()

fig.savefig(f'{graph_dir}price_pre.pdf')
```

```python
fig, ax = plt.subplots()
# fig.set_size_inches((16, 9))
ax.plot(x, y_rmse, label='mean absolute error', color=cs[2])

ax.set_xlabel('Time points')
ax.set_ylabel('MAE')
ax.legend()

fig.savefig(f'{graph_dir}online-mae.pdf')
```
